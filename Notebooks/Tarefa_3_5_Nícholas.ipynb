## PARTE I: *dataset [Advertising](https://github.com/malegopc/DSBD/blob/main/Datasets/Advertising/Advertising.csv)*

Este dataset mostra uma série de valores investidos em anúncios de TV, Rádio e jornais e os respectivos resultados de venda.

#### a-) Ler o dataset "*Advertising/Advertising.csv*" como dataframe utilizando a biblioteca Pandas e mostrar as 5 primeiras e 5 últimas linhas.

```
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
from scipy.stats import pearsonr

# Read Data Files
df = pd.read_csv('/content/DSBD/Datasets/Advertising/Advertising.csv')
# Eliminate null lines
df = df.dropna()
# Show Data
df[['TV', 'radio', 'newspaper', 'sales']]
```

#### b-) Mostre num mesmo quadro (usando subplots) os três diagramas de dispersão para os pares de variáveis: TV x sales, radio x sales e newspaper x sales.

```
TV = df['TV']
radio = df['radio']
newspaper = df['newspaper']
sales = df['sales']

plt.subplot(1,3,1) 
sbn.scatterplot(x=TV,y=sales)

plt.subplot(1,3,2) 
sbn.scatterplot(x=radio,y=sales)

plt.subplot(1,3,3) 
sbn.scatterplot(x=newspaper,y=sales)


plt.subplots_adjust(left   =    0,
                    bottom =    0, 
                    right  =  3.0, 
                    top    =  1.0, 
                    wspace = 0.25, 
                    hspace = 0.25)

plt.show()
```

#### c-) Calcule o coeficiente de correlação de Pearson para os três pares de variáveis: TV x sales, radio x sales e newspaper x sales.

```
# Pearson Correlation Coefficient TV x Sales
r, _ = pearsonr(TV, sales)
print('Coeficiente de correlação de Pearson TV x Sales:\t',r)

# Pearson Correlation Coefficient Radio x Sales
r, _ = pearsonr(radio, sales)
print('\nCoeficiente de correlação de Pearson Radio x Sales:\t',r)

# Pearson Correlation Coefficient Newspaper x Sales
r, _ = pearsonr(newspaper, sales)
print('\nCoeficiente de correlação de Pearson Newspaper x Sales:\t',r)

# Return Correlation Matrix
df[['TV', 'radio', 'newspaper', 'sales']].corr() 
```

#### d-) Analisando os diagramas de dispersão acima e os valores dos coeficientes de correlação responda qual dos tipos de anúncios influencia mais e qual deles influencia menos nos resultados das vendas?

Respostas: Atrvés dos dados obtidos acima, podemos dizer que a TV é o tipo de anúncio com maior influência no resultado das vendas, pois conseguimos verificar uma certa linearidade entre os dados do dataframe e a associação entre 'TV' e 'sales' possui o maior Coeficiente de Correlação Linear de Pearson.

Por outro lado, anúncios por meio de jornais é o tipo de anúncio que apresenta a menor influência nos resultados das vendas, pois apresentam uma baixa linearidade entre os dados do dataframe e a associação entre 'newspaper' e 'sales' possui o menor Coeficiente de Correlação Linear de Pearson.

## PARTE II: *dataset [Tips](https://github.com/mwaskom/seaborn-data/blob/master/tips.csv)*

Este dataset mostra diferentes atributos (valor da conta, gorjeta, número de pessoas, etc.) dos clientes de um restaurante.

#### e-) Faça a leitura do dataset "tips" disponível na biblioteca seaborn.

```
# Read Data Files
df = sbn.load_dataset('tips')
# Eliminate null lines
df = df.dropna()
# Show Data
df[['total_bill', 'tip', 'sex', 'smoker', 'day', 'time', 'size']]
```

#### f-) Mostre o diagrama de dispersão entre as variáveis "total_bill" e "tip" e calcule o coeficiente de correlação entr elas.





```
total_bill = df['total_bill']
tip = df['tip']

sbn.scatterplot(x=total_bill,y=tip)
plt.show()

# Pearson Correlation Coefficient
r, _ = pearsonr(total_bill, tip)
print('\nCoeficiente de correlação de Pearson:',r)

# Return Correlation Matrix
df[['total_bill', 'tip']].corr() 
```

ii-) Se sim, a correlação é positiva ou negativa? Resp: Positiva. 

iii-) Se sim, qual o grau de correlação? Forte, moderada ou fraca? Resp: É uma forte correlação positiva (0,60 < r < 1,00).

## PARTE III: *dataset [Atlas Brasil](https://raw.githubusercontent.com/malegopc/DSBD/main/Datasets/Atlas/Atlas_Brasil_2014.csv)*

Este dataset mostra diferentes atributos valores de variáveis sociais para cada um dos estados brasileiros:

* ANOSEST = Média de anos de estudo
* T_ANALF25M = Taxa de analfabetismo - 25 anos ou mais
* MORT1 =	Mortalidade infantil
* RDPC = Renda per capita 
* POPTOT = População total
* IDHM = IDHM

#### h-) Ler o dataset "*Atlas/Atlas_Brasil_2014.csv*" como dataframe utilizando a biblioteca Pandas

```
# Read Data Files
df = pd.read_csv('/content/DSBD/Datasets/Atlas/Atlas_Brasil_2014.csv')
# Eliminate null lines
df = df.dropna()
# Show Data
df[['ESTADO', 'MORT1', 'ANOSEST', 'T_ANALF25M', 'RDPC', 'POPT', 'IDHM']]
```

#### i-) Utilizando a função "[pairplot](https://seaborn.pydata.org/generated/seaborn.pairplot.html)" da biblioteca seaborn mostre os diagramas de dispersão com suas respectivas retas de regressão entre todos os pares de variáveis.

```
sbn.pairplot(df,kind='reg', dropna='boolean')
plt.show()
```

#### j-) Obtenha a matriz de correlação entre todos os pares de variáveis.

```
# Return Correlation Matrix
df[['MORT1', 'ANOSEST', 'T_ANALF25M', 'RDPC', 'POPT', 'IDHM']].corr() 
```

#### k-) Mostre um "mapa de calor" (heatmap) baseado na matriz de correlação entre todos os pares de variáveis.

```
sbn.heatmap(df.corr(), annot=True, cmap="BuPu")
plt.subplots_adjust(left=-1, wspace=1, hspace=-1, bottom=-0.5)
plt.show()
```
